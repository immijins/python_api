'소상공인시장진흥공단_상가(상권)정보_경북'

필요한 정보만 데이터 프레임으로 변경
경산, 표준산업분류명 : 커피 전문전 지도에 표시

```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_경북_202312.csv')
df.head()
```

```python
# 데이터 분류
df1 = df[(df.시군구명 == '경산시') & (df.표준산업분류명 == '커피 전문점')]
df2 = df1[['상호명', '시군구명', '경도', '위도']]
df2.tail()
```

```python
# 지도에 표시할 위도, 경도 위치
# 경산시 : 128.7415441, 35.8250555
gyeongsan_map = folium.Map(location=[35.8250, 128.7415], zoom_start=12)
gyeongsan_map
```

```python
# 지도 생성, 경산시의 커피 전문점
# zip : 오브젝트를 넣을 수 있음
for name, lat, lng in zip(df2['상호명'], df2['위도'], df2['경도']):
    folium.Marker([lat, lng], popup=name).add_to(gyeongsan_map)
gyeongsan_map
```

```python
# popup=name : 마우스 클릭 시 표시, tooltip=name : 마우스 오버시 표시
# Tooltip() permanent=True : 고정
for name, lat, lng in zip(df2['상호명'], df2['위도'], df2['경도']):
    # 마커 추가
    tooltip = folium.Tooltip(name, permanent=True)
    folium.Marker([lat, lng], tooltip=tooltip).add_to(gyeongsan_map)
gyeongsan_map
```
